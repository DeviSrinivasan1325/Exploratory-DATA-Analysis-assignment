## 2.Importing The Dataset

```python
raw_df = pd.read_csv("Zomato_Chennai_Dataset.csv")
raw_df.head()
```

## 3. Defining safe evaluation

```python
#to convert stringfield into python list:
def safe_eval(x):
    if isinstance(x, str):
        try:
            return ast.literal_eval(x)
        except:
            return []
    return []


list_cols = ['Cuisine', 'Top Dishes', 'Features']

for col in list_cols:
    raw_df[col] = raw_df[col].apply(safe_eval)
```

## 5. Cleaning The Dataset

## a. Removing the unnecessary columns or rows of data

```python
#Dropping unwanted columns and data
raw_df.drop(columns='Zomato URL',axis=1,inplace=True)
raw_df
```

```python
#Dropping Null value for better dataset
raw_df= raw_df.dropna(subset=['Delivery Rating','Dining Rating'])
raw_df
```

```python
#To check duplicate values in dataset
dupes_df = raw_df[raw_df.duplicated(
    subset=['Name of Restaurant', 'Location'],
    keep=False
)]
dupes_df.sort_values(['Name of Restaurant','Location'])
```

```python
raw_df = raw_df.drop_duplicates(
    subset=['Name of Restaurant', 'Location'],
    keep='first'
)
```

```python
raw_df
```

```python
raw_df.duplicated(
    subset=['Name of Restaurant', 'Location']
).sum()
```

```python
# Checking for Null records
raw_df.isnull().sum()
```

## b.Converting the DataTypes of numerical columns to numeric datatype

```python
raw_df.dtypes
```

```python
#Changing Data Type of the numerical columns
raw_df['Dining Rating Count'] = pd.to_numeric(raw_df['Dining Rating Count'], errors='coerce')
raw_df['Delivery Rating Count'] = pd.to_numeric(raw_df['Delivery Rating Count'], errors='coerce')
```

## Q3) Which are the Top 10 highest rated Biryani Restaurant in Chennai?

```python
#As we used the safe evaluation to convert top dishes into python list, we cant use the list to check as string. so i convert that back to string and maintained as Top Dishes_str
for col in ['Top Dishes']:
    zomato_df['Top Dishes_str'] = zomato_df[col].apply(lambda x: ', '.join(x) if isinstance(x, list) else '')
```

```python
biryani_top10_fixed = (
    biryani_df
    .groupby('Name of Restaurant', as_index=False)
    .agg({
        'Dining Rating': 'max',
        'Location': 'first'
    })
    .sort_values('Dining Rating', ascending=False)
    .head(10)
)
biryani_top10_fixed
```

```python
fig1= px.bar(biryani_top10_fixed,x='Dining Rating', y='Name of Restaurant', orientation='h',
    title='Top 10 Highest-Rated Biryani Restaurants in Chennai'
)
fig1.show()
```

## Q4) Which restaurants in Chennai have the highest ratings (Dining or Delivery ≥ 4.5) for each Cuisine Type, and in which localities are they located?

```python
condition = (
    (zomato_df['Dining Rating'] >= 4.5) |
    (zomato_df['Delivery Rating'] >= 4.5)
)

highest_rated_df = zomato_df[condition]
highest_rated_df = highest_rated_df.explode('Cuisine')

highest_rated_df['Max Rating'] = highest_rated_df[['Dining Rating', 'Delivery Rating']].max(axis=1)

result = highest_rated_df.loc[highest_rated_df.groupby('Cuisine')['Max Rating'].idxmax(), 
                              ['Cuisine', 'Name of Restaurant', 'Location', 'Max Rating']]
result = result.reset_index(drop=True)
result
```

```python
fig = px.bar(
    result,
    y='Cuisine',          # swap axes
    x='Max Rating',
    color='Cuisine',
    text='Name of Restaurant',
    orientation='h',      # horizontal bars
    hover_data=['Location', 'Max Rating'],
    title='Highest Rated Restaurants per Cuisine in Chennai (Rating ≥ 4.5)'
)

fig.update_traces(textposition='inside', textfont_size=10)
fig.update_layout(yaxis={'categoryorder':'total ascending'}, showlegend=False)
fig.show()
```

## Q5) What is the Avg Price Distibution of highest rated restaurant for each Cuisine Type in Chennai?

```python
highest_rated_price_df = pd.merge(
   result, 
    zomato_df[['Name of Restaurant', 'Price for 2']],  # make sure this column exists in zomato_df
    on='Name of Restaurant',
    how='left'
)

# Now you can group by Location and Cuisine
avg_price_df = highest_rated_price_df.groupby(['Location', 'Cuisine'])['Price for 2'].mean().reset_index()
avg_price_df
```

```python
fig = px.scatter(highest_rated_price_df,x='Location',y='Price for 2', color='Cuisine',symbol='Cuisine',
                 title='Avg Price Distribution of High rated restaurant for each Cusine Type').update_traces(marker_size=10)
fig.show()
                 
```

## Q6) Which areas have a large number of Chinese Restaurant Market?

```python
# If 'Cuisine' is a list
zomato_df['Cuisine'] = zomato_df['Cuisine'].apply(lambda x: x if isinstance(x, list) else [x])

# Filter Chinese restaurants
chinese_df = zomato_df[zomato_df['Cuisine'].apply(lambda x: 'Chinese' in x)]

# Count per Location
chinese_count_by_area = chinese_df.groupby('Location')['Name of Restaurant'].count().reset_index()
chinese_count_by_area.rename(columns={'Name of Restaurant': 'Chinese_Restaurant_Count'}, inplace=True)

# Sort descending to find areas with the largest market
chinese_count_by_area = chinese_count_by_area.sort_values(by='Chinese_Restaurant_Count', ascending=False)
chinese_count_by_area=chinese_count_by_area.head(10)
chinese_count_by_area
```

```python
fig = px.bar(chinese_count_by_area, x='Location', y='Chinese_Restaurant_Count', color='Location', title= 'No. of Chinese Restaurant by places')
fig.show()
```

## Q9) Find the list of Affordable Restaurants?

## The criteria for Affordable Restaurants would be:-
## 1) Low Price 2) High Rated
## First step will be to find the restaurants with average cost 1/4th the average cost of most expensive restaurant in our dataframe.
## Let me explain:-The most expensive restaurant has an average meal cost= 6000. We'll try to stay economical and only pick the
## restaurants that are 1/4th of 6000

```python
max_price = zomato_df['Price for 2'].max()
one_fourth_price = max_price/4
one_fourth_price
```

```python
aff_rest_df = zomato_df[['Name of Restaurant','Price for 2','Cuisine','Location']]
aff_rest_df= aff_rest_df[aff_rest_df['Price for 2']<=750]
aff_rest_df.sort_values(by='Price for 2',inplace=True)
aff_rest_df
```

```python
highrate_rest_df=zomato_df[['Name of Restaurant','Price for 2','Cuisine','Location', 'Dining Rating']]
highrate_rest_df = highrate_rest_df[highrate_rest_df['Dining Rating']>=4.5]
highrate_rest_df.sort_values(by='Price for 2', inplace=True)
highrate_rest_df
```

```python
#merge both aff_rest_df and highrate_aff_df
```

```python
highrate_aff_df= pd.merge(aff_rest_df,highrate_rest_df, how='inner',on=['Name of Restaurant','Location'])
highrate_aff_df= highrate_aff_df[['Name of Restaurant','Price for 2_x','Cuisine_x','Location']]
highrate_aff_df.rename(columns={'Name of Restaurant':'Name of Restaurant','Price for 2_x':'Price','Cuisine_x':'Cuisine',
                                'Location':'Location'},inplace=True)
highrate_aff_df
```
